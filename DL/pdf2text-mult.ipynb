```python
import os
from pdf2image import convert_from_path
import pytesseract
from joblib import Parallel, delayed
import multiprocessing
from tqdm import tqdm
```

```python
def raspoz(fnn):
    if fnn[-3:]=='pdf':
        try:
            if os.path.exists('./txt/txt__'+fnn.replace('/', '_').replace(' ', '_')[2:-4]+'__0000.txt')==False:
                pagesPDF = convert_from_path(fnn, 300)
                doci=0
                for pagePDF in pagesPDF:
                    print(doci)
                    classF=fnn.split('/')[-1]
                    #pagePDF.save('./img/out_'+classF+'_'+str(doci)+'.jpg', 'JPEG')
                    if os.path.exists('./txt/txt__'+fnn.replace('/', '_').replace(' ', '_')[2:-4]+'__'+str(doci).zfill(4)+'.txt')==False:
                        pageTEXT=pytesseract.image_to_string(pagePDF,lang='rus')                
                        with open('./txt/txt__'+fnn.replace('/', '_').replace(' ', '_')[2:-4]+'__'+str(doci).zfill(4)+'.txt', "w") as output:
                            output.write(pageTEXT)                    
                        doci=doci+1
        finally:
            print(fnn)
            
    
```

```python
file_list=[]
for top, dirs, files in os.walk('./Dataset'):
    for nm in files:  
        fnn=os.path.join(top, nm)
        file_list.append(fnn)
        #raspoz(top,nm)
file_list
```

```python
results = Parallel(n_jobs=6)(delayed(raspoz)(i) for i in tqdm(file_list))
print(results)
```
